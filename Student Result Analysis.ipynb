```python
 df = pd.read_csv("student.csv")
df.isnull().sum()
```

# Drop Unknown Column


```python
df = df.drop("Unnamed: 0",axis =1)
print(df.head())
```

# Change weekly study hours column

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("05-oct","5-10")
df.head()
```

# gender distribution

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(5,5))
ax = sns.countplot(data=df, x="Gender")  # Assign plot to 'ax'
ax.bar_label(ax.containers[0])           # Use ax.containers[0] (not ax.container[0])
plt.show()
```

```python
gb = df.groupby("ParentEduc").agg({
    "MathScore": "mean",
    "ReadingScore": "mean",
    "WritingScore": "mean"
})

print(gp)
```

```python
plt.figure(figsize=(3,3))
sns.heatmap(gb, annot = True)
plt.title("Relationship between parent's education and students score")
plt.show()
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({
    "MathScore": "mean",
    "ReadingScore": "mean",
    "WritingScore": "mean"
})

print(gb1)
```

```python
plt.figure(figsize=(3,3))
sns.heatmap(gb1, annot = True)
plt.title("Relationship between Parent's Marital Status on Student Score")
plt.show()
```

# Distribution of Ethnic Group

```python
groupA = df.loc[(df["EthnicGroup"] == "group A")].count()
print(groupA)
```

```python
groupA = df.loc[(df["EthnicGroup"] == "group A")].count()
groupB = df.loc[(df["EthnicGroup"] == "group B")].count()
groupC = df.loc[(df["EthnicGroup"] == "group C")].count()
groupD = df.loc[(df["EthnicGroup"] == "group D")].count()
groupE = df.loc[(df["EthnicGroup"] == "group E")].count()
l = ["group A","group B", "group C", "group D","group E"]
mlist =[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
print(mlist)
plt.pie(mlist,labels = l , autopct ="%1.2f%%")
plt.title("Distribution of Ethnic Group")
plt.show()
```
